```python
x =d.groupby("Company")["Price (Euro)"].sum().sort_values(ascending=False)
c1 = x.values
c2 = x.index
print(c1)
print(c2)
```

```python
plt.plot(c2, c1)
plt.xticks(rotation=90)
plt.show()
```
